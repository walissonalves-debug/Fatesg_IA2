```python
import pandas as pd

df_loja1 = pd.DataFrame({
    'produto': ['Mouse', 'Teclado'],
    'Estoque': [20,15]
})

df_loja2 = pd.DataFrame({
    'produto': ['Monitor', 'Fone'],
    'Estoque': [1000,150]
})
```

```python
df_total = pd.concat([df_loja1 , df_loja2], ignore_index = True)

print("Resultado da concatenação:")
print(df_total)
```
