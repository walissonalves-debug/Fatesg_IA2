```python
import statistics as st
def notas():
    n1 = float(input("Informe sua N1: "))
    n2 = float(input("Informe sua N2: "))
    n3 = float(input("Informe sua N3: "))
    
    entradas = (n1,n2,n3)
    media = round(st.mean(entradas),2)

    print(f'Sua n1 é {n1}, sua n2 é {n2} e sua n3 é {n3}. Sua média é {media}.')


    if media< 4:
        print("Você está reprovado(a)!")
    elif media < 6:
        print("Você está de recuperação: ")
    else:
        print("Você está aprovado")
```

```python
notas()
```
